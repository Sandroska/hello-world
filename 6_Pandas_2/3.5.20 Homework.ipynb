2. Write one function that can accomplish any of the three cases from the previous point by specifying extra parameters in the function call. Here are three examples of how you would use each function to clean a column: 
    - `ga["Users"] = clean_function(ga["Users"], ',', 'int64')`

    - `ga["Revenue"] = clean_function(ga["Revenue"], 'US$', 'float64')`

```python
# define the function clean_function()
def clean_function(series, token, types):
  try:
    clean = series.str.replace(token, '', regex = False).astype(types)
  except:
    return series
  return clean  
```

```python
# call the function clean_function()
# first to reset the data
ga = pd.read_csv("/content/gdrive/MyDrive/Sander Mulder/python folder/Pandas 2/data/GA Paid Search Traffic.csv", skiprows=[0,1,2,3,4,5,8], nrows=10)
# call the functions

ga["Users"] = clean_function(ga["Users"], ',', 'int64')
ga["Bounce Rate"] = clean_function(ga["Bounce Rate"], '%', 'float64')
ga["Revenue"] = clean_function(ga["Revenue"], 'US$', 'float64')
```
